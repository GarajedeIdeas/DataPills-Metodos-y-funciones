# Métodos vs Funciones en Python


Python es un lenguaje multiparadigma, que soporta **programación orientada a objetos (POO)**, siendo las clases y los objetos sus principales bloques de desarrollo. Las clases son los modelos sobre los cuáles se construirán nuestros objetos, mientras que los objetos son instancias de esa clase.



```python
import pandas as pd
```

```python
string = 'esto es un string'
lista = [10,3,7]
df = pd.DataFrame({'col1': [1, 2], 'col2': [3, 4]})
```

## Métodos

Para utilizar un médodo, nombramos el objeto, seguido de un punto (.) y escribimos el método a usar

```python

objeto.método(argumentos_del_método)
```

```python
# Métodos asociados a strings
print(string.upper())
print(string.replace('i','o'))
```

```python
# Métodos asociados a listas
lista.append(4)
lista.sort()
lista
```

```python
# Las clases pueden contener métodos y atributos
df.columns
```

```python
# Métodos asociados a dataframes
df.set_index('col1')
```

```python
df.rename(columns={'col1':'x',
                   'col2':'y'})
```

Para acceder a los métodos asociados a un objeto, podemos utilizar la función `dir()`

```python
dir(df)
```

## Funciones

Para utilizar un médodo, nombramos el objeto, seguido de un punto (.) y escribimos el método a usar

```python
funcion(argumentos_de_funcion)
```

```python
# Built-in functions
print(len(string))
print(len(lista))
print(len(df))
```

```python
# Funciones de terceros
import math
for x in lista:
    print(math.sqrt(x))
```

```python
# User-defined functions (UDF)
def wordcount(texto):
    palabras = texto.split()
    num_palabras = len(palabras)
    return num_palabras
```

```python
wordcount(string)
```
